# Кортеж


Кортеж состоит из ряда значений, разделенных запятыми, наприм

```python
tpl1: tuple[int, int, str] = 1234, 5678, "xyz"
# здесь переменная t - это кортеж
print(tpl1)
```

```python
# 20
tpl2: tuple[tuple[int, int, str], tuple[str, ...]] = tpl1, ("a", "b", "c")
print(tpl2)
```


Кортежи похожи на списки, но с двумя отличиями.<br>
*  Кортежи неизменяемы, а списки изменяемы.<br>
*  Кортежи заключаются в круглые скобки, а списки — в квадратные<br>
* Кортежи неизменяемы и обычно содержат неоднородную последовательность<br>
элементов, доступ к которым осуществляется посредством распаковки или ин­<br>
дексации.<br>
* Почти все методы списков также применимы и к кортежам. Вы уже знаете, что<br>
кортежи неизменяемы. Следовательно, методы и функции, которые изменяют спи­<br>
ски, для кортежей использоваться не могут.

# Словарь


Словарь— это еще один встроенный тип данных Python. Это неупорядоченный<br>
набор единиц данных.<br>
В обычных словарях, как правило, есть ключ, т. е. слово или элемент, значение ко­<br>
торого объясняется в словаре. Это объяснение и есть значение ключа.Пустой словарь созда­<br>

пары ключ-, значение.<br>
Словарь также можно создать с помощью встроенной функции<br>
diet о1.<br>
Ключ в словаре — это уникальный и неизменяемый тип данных, такой как строка,<br>
число или кортеж. Другими словами, в качестве ключей можно использовать толь­<br>
ко простые типы данных.

```python
print(dict([("заре", 4139), ("guide", 4127), ("jack", 4098)]))
```

```python
print({x_: x_**2 for x_ in (2, 3, 4, 5)})
```


Перебор последовательностей в цикле

```python
knights: dict[str, str] = {"Gallahad": "the pure", "Robin": "the brave"}
for k_, v_ in knights.items():
    print(k_, v_)
```

```python
for i_, v_ in enumerate(["tic", "tac", "toe"]):
    print(i_, v_)
```

```python
questions: list[str] = ["name", "quest"]
answers: list[str] = ["Lancelot", "favorite color"]
```

```python
for q_, a_ in zip(questions, answers):
    print(f"What is your {q_}? It is {a_}")
```

```python
# обратный порядок
for i_ in reversed(range(1, 10, 2)):
    print(i_)
```


перебрать последовательность в отсортированном порядке

```python
basket: list[str] = ["apple", "orange", "apple", "pear", "orange", "banana"]
for f_ in sorted(set(basket)):
    print(f_)
```

```python
# 30
lst_0: list[float] = [56.2, float("NaN"), 51.7]
lst_1: list[float] = [55.3, 52.5, float("NaN"), 47.8]
raw_data: list[float] = lst_0 + lst_1
filtered_data: list[float] = []
for value in raw_data:
    if not math.isnan(value):
        filtered_data.append(value)
print(filtered_data)
```

   - Списки являются изменяемыми (mutable), что означает, что их содержимое можно изменять после создания (например, добавлять или удалять элементы).    - Кортежи сохраняют порядок элементов, что позволяет обращаться к ним по индексу. Множества не имеют фиксированного порядка, и элементы в них не индексируются. Это влияет на использование: кортежи подходят для хранения связанных данных, где важен порядок, тогда как множества хороши для хранения уникальных элементов и выполнения операций над множествами (например, объединение).<br>

### 8.8.2. Правда или ложь<br>
<br>
1. **Ложь** - Строки в Python неизменяемы (immutable), поэтому при изменении строки создается новая строка в памяти.<br>
   <br>
2. **Правда** - Операция умножения строки на число возвращает строку с повторяющимися значениями (например, 'abc' * 3 даст 'abcabcabc').<br>
<br>
3. **Ложь** - Строки не различаются в зависимости от используемых кавычек (одинарные или двойные кавычки могут использоваться взаимозаменяемо).<br>
<br>
4. **Правда** - Индексирование действительно работает со строками и списками, но не со множествами, так как множества не имеют порядка.<br>
<br>
5. **Ложь** - В списках может храниться несколько типов данных одновременно (например, смешанные типы).<br>
<br>
6. **Правда** - Вставить новое значение в кортеж нельзя, так как кортежи неизменяемы.<br>
<br>
7. **Правда** - Списковые включения действительно являются элегантным методом управления списками.<br>
<br>
8. **Правда** - Кортеж может содержать разные типы данных (например, (1, "text", 3.14)).<br>
<br>
9. **Правда** - Кортежи заключаются в круглые скобки (), а списки — в квадратные [].<br>
<br>
10. **Правда** - В словаре количество ключей и значений не обязательно должно быть одинаковым; ключи могут быть уникальными, а значения могут повторяться.


6. Функция для проверки отсортированного списка

```python
def is_sorted_list(lst: list[int]) -> bool:
    """Проверяет, отсортирован ли список.

    Аргументы:
        lst (list): Список для проверки.

    Возвращает:
        bool: True сортировка по возрастанию,
            иначе False.
    """
    return lst == sorted(lst)
```

Пример использования:

```python
print(is_sorted_list([1, 2, 3, 4]))  # True
print(is_sorted_list([3, 2, 1]))  # False
```


8. Функция для удаления дубликатов из списка

```python
def skip_duplicates(lst: list[int]) -> list[int]:
    """Удаляет дубликаты из списка.

    Аргументы:
        lst (list): Исходный список.

    Возвращает:
        list: Список без дубликатов.
    """
    return list(set(lst))
```

Пример использования:

```python
orign_list: list[int] = [1, 2, 2, 3, 4, 4]
unique_list: list[int] = skip_duplicates(orign_list)
print(unique_list)  # [1, 2, 3, 4]
```


9. Функция для вычисления произведения элементов списка

```python
def product_of_list(lst: list[int]) -> int:
    """Вычисляет произведение элементов списка.

    Аргументы:
        lst (list): Список чисел.

    Возвращает:
        int: Произведение элементов списка.
    """
    return reduce(lambda x_, y_: x_ * y_, lst)
```

Пример использования:

```python
numbers: list[int] = [1, 2, 3, 4]
print(product_of_list(numbers))  # 24
```
