
Цикл while выполняется до тех пор, пока условие остается истинным (в первом<br>
варианте это было а < 10, которое мы позже изменили на а < 15). В Python ис­<br>
тинным является любое ненулевое целочисленное значение, ноль считается<br>
ложным. Условие также может быть строкой или списком и вообще любой по­<br>
следовательностью. При этом все, что имеет ненулевую длину, считается истин­<br>
ным, а пустые последовательности — ложными. Когда составное выражение вводится в интерак­<br>
тивном режиме, за ним должна следовать пустая строка, указывающая на завер­<br>
шение (поскольку синтаксический анализатор не может угадать, когда блок кода<br>
закончился). Обратите внимание, что у всех строк в блоке должен быть одина­<br>
ковый отступ.

две важные концепции программи­<br>
рования на Python:<br>
♦ Аргументы (иногда обозначаемые словом<br>
args);<br>
♦ Именованные аргументы (иногда обозначаемые словом<br>
kwargs).


Аргументы — это все, что мы передаем в функцию. Например, в функцию print о<br>
мы передавали строку или переменную — это и есть аргументы. Если при опреде­<br>
лении функции вы хотите передавать ей больше одной переменной, но еще не знае­<br>
те точно, сколько аргументов у вас будет, можно указать специальный синтаксис<br>
*args. Символ * здесь означает, что мы можем передать функции любое количество<br>
переменных, а слово args — это своего рода стандартное соглашение, которое ис­<br>
пользуется для лучшей читаемости кода. Можно использовать и любое другое сло­<br>
во, если оно будет достаточно значимым. Как видите, в синтаксисе функции print ()<br>
вместо *args используется слово ‘objects. Теперь вы знаете, что это означает,<br>
и в функцию print!) можно передать несколько аргументов (переменных, строк,<br>
классов или любых других объектов).


Именованные аргументы — это аргументы, для которых при передаче в функцию<br>
указывается не только значение, но и имя.


давайте остановимся и рассмот­<br>
рим подробнее это понятие на пример синтаксиса функции print!).<br>
Вы заметили в заголовке функции слова sep=, end=, file= и fiush=? Это именованные<br>
аргументы, у них есть имя, а справа от символа = указано присвоенное им значение.<br>
Значения, указанные в определении функции, являются значениями по умолчанию.<br>
То есть если вы не назначите такому аргументу другое значение, будет передано<br>
значение по умолчанию, и синтаксических ошибок не возникнет.


Параметры функции print ():<br>
* ‘objects — любое количество любых объектов. Перед выводом все объекты со­<br>
бираются в строку;<br>
* sep=' ' (необязательный)— задает разделитель объектов, если их несколько.<br>
Значение по умолчанию — символ пробела, указанный как ' ';<br>
* end='\n' (необязательный)— определяет, что вывести в конце строки. Значение<br>
по умолчанию — символ переноса строки '\п';<br>
* **file=sys. stdout** (необязательный) — объект с методом write (string). Если этот<br>
параметр не указан, будет использоваться sys. stdout по умолчанию, что означает<br>
вывод результатов на экран;<br>
* **fiush=False** (необязательный) — логическое значение, указывающее, будет ли<br>
вывод очищен (True) или буферизован (False). По умолчанию имеет значение<br>
False.


**Рассмотрим три основных типа ошибок, которые возникают чаще всего.**<br>
* 1) Синтаксическая ошибка — проблемы с языковыми конструкциями: SyntaxError: ...<br>
Лучшее решение, чтобы избежать длительного поиска ошибок, — это запускать<br>
программу через каждые несколько строк кода.<br>
* 2) Ошибка времени выполнения — возникает, когда Python понимает саму команду, но<br>
при ее выполнении сталкивается с проблемами. Поэтому эта ошибка и называется<br>
«времени выполнения», поскольку возникает только после запуска программы.<br>
   * неверно введенное имя переменной или функции;<br>
   * использование переменной до ее определения;<br>
   * имя должно было быть заключено в кавычки;<br>
   * деление на ноль.<br>
* 3) Семантическая ошибка — это проблемы с самим построением<br>
вашей программы. Обычно такие проблемы не вызывают сообщений об ошибках,<br>
но поведение программы оказывается неверным. Ошибки такого типа сложнее всего<br>
отследить. часто бывают вызваны случайным использованием неверных пере­<br>
менных либо просто неправильными вычислениями.
