```python
# решение задачи с помощью циклов. Подумаем заново:
# ♦ один цикл будет печатать последовательно 6 рядов;
# ♦ другой цикл будет печатать сами звездочки в рядах.
# row =6 # количество строк или высота треугольника


def pattern_t(row: int) -> str:
    """Print triangle."""
    for i_ in range(0, row):  # цикл для перебора строк
        for _ in range(0, i_ + 1):  # цикл для перебора в диапазоне
            # от 0 до 1 + номер строки
            print("* ", end="")  # вывод звездочек в строке
        print("")  # переход на следующую строку


pattern_t(6)
```

**Перевернутый прямоугольный треугольник**

```python
def pattern_t_inverse(num: int) -> str:
    """Print triangle."""
    spaces = (num - 1) * 2
    for i_ in range(0, num):
        for j_ in range(0, spaces):
            print(" ", end="")
        for j_ in range(0, i_ + 1):
            print("* ", end="")
        spaces = spaces - 2
        print("")


pattern_t_inverse(6)
```

**Равносторонний треугольник**

```python
def pattern_et(num: int) -> str:
    """Print equilateral triangle."""
    spaces: int = num - 1
    for row in range(0, num):
        for space in range(0, spaces):
            print(" ", end="")
        for star in range(0, row + 1):
            print("* ", end="")
        spaces = spaces - 1
        print("")


pattern_et(6)
```

```python
# Picture 1


def pattern_riversed_et(num: int) -> str:
    """Print riversed equilateral triangle."""
    spaces: int = 1

    for row in range(num, 0, -1):
        for space in range(0, spaces):
            print(" ", end="")

        for star in range(0, row):
            print("* ", end="")

        spaces += 1
        print("")


pattern_riversed_et(6)
```

```python
# Picture 6


def triangle(row: int) -> str:
    """Print two triangles."""
    # Верхний треугольник
    for i_ in range(row, 0, -1):  # от row до 0 в обратном порядке
        # Вывод пробелов
        for space in range(row - i_):
            print(" ", end="")

        # Вывод звездочек
        for star in range(i_):
            print("* ", end="")

        print("")  # Переход на новую строку

    # Нижний треугольник
    for i_ in range(1, row + 1):  # проходим ряд от 1 до row+1
        # Вывод пробелов
        for space in range(row - i_):  # к-во пробелов от ряд - i
            print(" ", end="")

        # Вывод звездочек
        for star in range(i_):  #
            print("* ", end="")

        print("")  # Переход на новую строку


triangle(6)
```

**Прямоугольный треугольник вверх ногами**

```python
def numpattern3(number: int) -> int:
    """Print a right-angled triangle upside down."""
    num = 1
    for i_ in range(number, 0, -1):
        for j_ in range(0, i_):
            print(number, end=" ")
        num = number + 1
        print("")


numpattern3(6)
```

**Ромб**

```python
def numpattern6(num: int) -> None:
    """Prints a number pattern based on the input integer.

    Args:
        num (int): The height of the number pattern.
    """
    spaces = num - 1
    num_count = 1

    # Upper part of the pattern
    for row in range(1, num + 1):
        for _ in range(spaces):
            print(" ", end="")
        for _ in range(num_count):
            print(row, end=" ")
        print("")
        spaces -= 1
        num_count += 1

    spaces = 1
    num_count = num - 1

    # Lower part of the pattern
    for row in range(num - 1, 0, -1):
        for _ in range(spaces):
            print(" ", end="")
        for _ in range(num_count):
            print(row, end=" ")
        print("")
        spaces += 1
        num_count -= 1


# Example usage
numpattern6(6)
```

**Ромб с другим направлением чисел**

```python
def numpattern7(num: int) -> None:
    """Prints rhombus with a different direction of numbers.

    Args:
        num (int): The height of the number pattern.
    """

    spaces = num
    num_count = 1
    for row in range(0, num + 1):
        for i_ in range(0, spaces):
            print(" ", end="")
        for num in range(1, num_count):
            print(num, end=" ")
        print("")
        spaces = spaces - 1
        num_count = num_count + 1

    spaces = 1
    num_count = num
    for row in range(num, 0, -1):
        for i in range(0, spaces):
            print(" ", end="")
        for num in range(1, num_count):
            print(num, end=" ")
        spaces = spaces + 1
        num_count = num_count - 1
        print("")


numpattern7(6)
```
