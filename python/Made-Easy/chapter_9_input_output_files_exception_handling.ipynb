# Чтение и запись файлов

Обычно файлы открываются в текстовом режиме, что означает, что вы читаете из
файла и записываете в файл строки в определенной кодировке.
Метод open о возвращает файловый объект и чаще всего используется с двумя
аргументами:
f = open (имя_файла, режим')
Первый аргумент— это строка, содержащая имя файла. Второй аргумент— это
тоже строка-режим использования
файла.

Режим может состоять из:
* ♦ 'г' —  только для чтения; СТОИТ ПО УМОЛЧАНИЮ
* ♦'w' — только для записи (если файл с таким же именем уже
существует, он будет перезаписан);
* ♦ ' а' —  для добавления; любые данные, записываемые в файл,
добавляются в его конец;
* ♦ ' г+ ’ — файл открывается как для чтения, так и для записи.

При работе с файловыми объектами рекомендуется использовать ключевое слово

ключение.

Если вы не используете ключевое слово with, то в конце нужно вызвать метод
f.close(), чтобы закрыть файл и высвободить все используемые им системные
ресурсы.

Если достигнут конец файла, f.read() вернет пустую строку (''), если файл состоит из одной строки, то "". Метод f. readline () читает одну строку из файла.При каждом вызове метода читается следующая строка.

# Создание файла и запись в него

```python
# 15
# строковая переменная с произвольным текстом
story: str = (
    "Let's write a small story.n"
    "There was a curious student. He wanted to learn Python.n"
    "His age was... Well, how does it matter? One can be a student at any age.n"
    "Let's say, he was a good student and within a short time, learnt all aboutn"
    "programming and Python. He combined that with his business acumen andn"
    "became a good data scientist."
)

# создание нового текстового файла
file_name: str = "small_story.txt"

# Выполнив последнюю строку кода, мы создаем новый файл с именем small_story.txt.
with open(file_name, "w") as f:
    # запись текстовой переменной story в файл f
    f.write(story)
    # получаем количество символов, записанных в файле
    char_count: int = len(story)

# char_count можно использовать для дальнейшей обработки, если необходимо
print(f"Записано {char_count} символов в файл '{file_name}'.")
```

Давайте по порядку ответим на ваши вопросы.

### 1. Что делает функция input()?
Функция input() в Python используется для считывания строки, введенной пользователем с клавиатуры. Она останавливает выполнение программы и ждет, пока пользователь введет данные и нажмет Enter. В Python 3.x input() всегда возвращает строку. В Python 2.x для получения строки используется raw_input(), а input() пытается выполнить введенное выражение.

### 2. Как отформатировать строки с помощью метода format()?
Метод format() позволяет вставлять значения в строки, используя фигурные скобки {} как плейсхолдеры. Например:
name = "Alice"
age = 30
formatted_string = "Name: {}, Age: {}".format(name, age)
print(formatted_string)  # Вывод: Name: Alice, Age: 30

Можно также использовать именованные параметры:
formatted_string = "Name: {name}, Age: {age}".format(name=name, age=age)


### 3. Как ограничить размер считываемого файла на Python? Почему это важно?
Чтобы ограничить размер считываемого файла, можно использовать аргументы в методах чтения, например, при чтении с помощью read(size) можно указать количество байт для чтения. Это важно для управления памятью и предотвращения загрузки слишком больших файлов в оперативную память, что может привести к замедлению работы программы или ее падению.

### 4. Зачем нам нужно открывать сторонние файлы в Python?
Открытие сторонних файлов позволяет программе взаимодействовать с внешними данными, такими как текстовые документы, изображения, базы данных и т.д. Это дает возможность считывать данные для обработки или записывать результаты работы программы в файл для дальнейшего использования.

### 6. Сравните работу методов write() и writelines().
- **write()**: записывает одну строку в файл. Если вы хотите записать несколько строк, нужно вызывать его несколько раз.
- **writelines()**: принимает список строк и записывает их все сразу в файл. Однако он не добавляет символы новой строки между строками, поэтому их нужно добавлять вручную.

Пример:
with open('example.txt', 'w') as f:
    f.write("Hello\n")
    f.writelines(["Line 1\n", "Line 2\n"])


### 7. Сравните методы read(), readline() и readlines().
- **read()**: считывает весь файл целиком и возвращает его содержимое как одну строку.
- **readline()**: считывает одну строку за раз из файла. Каждый вызов возвращает следующую строку.
- **readlines()**: считывает все строки файла и возвращает их в виде списка строк.

### 8. Какие режимы можно использовать с функцией open() при открытии файла?
Некоторые основные режимы:
- 'r': чтение (по умолчанию).
- 'w': запись (создает новый файл или перезаписывает существующий).
- 'a': добавление (добавляет данные в конец файла).
- 'b': бинарный режим (например, 'rb' для чтения бинарных файлов).
- 't': текстовый режим (по умолчанию).

Пример открытия файла для записи:
with open('example.txt', 'w') as f:
    f.write("Hello")


### 9. Объясните работу методов str() и repr().
- **str()**: возвращает строковое представление объекта, которое удобно для человека (читаемое).
- **repr()**: возвращает строковое представление объекта, которое удобно для разработчика и может быть использовано для восстановления объекта (если это возможно).

Пример:
s = "Hello"
print(str(s))   # Вывод: Hello
print(repr(s))  # Вывод: 'Hello'

3.Напишите функцию, которая принимает три аргумента: входной файл, выход­ной файл и функцию преобразования. Первый аргумент - это файл, открытый для чтения, второй аргумент - это файл, открытый для записи, а третий аргу­мент - это функция преобразования, которая принимает строку на вход, вы­полняет какое-то преобразование по вашему выбору и возвращает измененную строку. Основная функция должна читать строку из входного файла, выполнять над ней функцию преобразования, а затем записывать измененную строку в вы­ходной файл. Преобразование может быть, например, таким: каждое слово в строке должно начинаться с заглавной буквы.

```python
def case_change(input_: str, output: str, method: str = "up") -> str:
    """_summary_.

    Args:
        input_ (str): _description_
        output (str): _description_
        method (str, optional): _description_. Defaults to "up".

    Returns:
        str: _description_
    """
    with open(input_, encoding="utf-8") as file1:
        string1: str = file1.read()
    with open(output, "w", encoding="utf-8") as file2:
        if method == "up":
            file2.write(string1.upper())
        else:
            file2.write(string1.lower())

    return "the text is annotated in " + output
```

```python
def htg_in_file(file_path: str) -> list[str]:
    """_summary_.

    Args:
        file_path (str): _description_

    Returns:
        list[str]: _description_
    """
    with open(file_path, encoding="utf-8") as f_:
        final_list: list[str] = []
        for string in f_.readlines():
            if "#" in string:
                final_list.append(string)
    return final_list
```

6.Читать файл с начала - это просто, но что если вам нужно прочитать файл в обратном направлении, например для чтения файлов с логами. Напишите про­грамму для чтения и отображения содержимого файла от конца до начала.

```python
def reverse_file(path: str) -> str:
    """_summary_.

    Args:
        path (str): _description_

    Returns:
        str: _description_
    """
    with open(path, encoding="utf-8") as f_:
        reved = f_.read()[::-1]
    return reved
```
