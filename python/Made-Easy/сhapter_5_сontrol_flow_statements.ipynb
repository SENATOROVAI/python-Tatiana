Оператор break прерывает выполнение самого внутреннего цикла for или while.
Операторы цикла могут иметь блок else, который выполняется, когда цикл завер­
шается, перебрав все элементы (в случае цикла for) или когда условие становится
ложным (в случае цикла while). Обратите внимание, что блок else не выполняется,
когда цикл завершается оператором break (рис. 5.8)

image.png

```python
numbers: list[int] = [1, 2, 3, 4, 5]
for i_ in numbers:
    print(i_)
    break
else:
    print("this is end of the code!")
print("this line is outside the loop")
# оператор break остановил цикл for и не перевел его на
# следующую итерацию и про­пустил блок else.
```

```python
for n_ in range(2, 10):
    for x_ in range(2, n_):
        if n_ % x_ == 0:
            print(n_, " equals ", x_, "*", n_ // x_)
            break
    else:
        print(n_, "is а prime number ")
```

# 5.9. Упражнения

1. Каков порядок выполнения выражений в программе Python? Как его изменить?




* Скобки: Выражения внутри скобок вычисляются первыми.

* Возведение в степень: Операции возведения в степень (`) выполняются в порядке следования слева направо.

* Унарные операторы: Унарные операторы, такие как +, -, ~, выполняются после операций со скобками и возведения в степень.

* Умножение, деление, остаток, целочисленное деление: Эти операции выполняются в порядке следования слева направо.

* Сложение и вычитание: Операции сложения и вычитания выполняются в порядке следования слева направо.

* Операторы сравнения: Операторы сравнения <, >, <=, >=, ==, != выполняются после арифметических операций.

* Логические операторы: Логические операции not, and, or выполняются после операторов сравнения.

* Операторы присваивания: Операторы присваивания, такие как =, +=, -=, *=, /=, выполняются в последнюю очередь.

* Чтобы изменить порядок выполнения выражений, вы можете использовать круглые скобки (). Выражения внутри скобок будут вычисляться в первую очередь. Это позволяет переопределить стандартный порядок выполнения операций.

```python
# 3. Трое сотрудников, Самир, Правин и Мохит, вводят свой опыт работы с клавиа­
# туры. Напишите программу, которая определяет наиболее и наименее опытного
# из них.

# Ввод опыта работы сотрудников
samir_experience: int = int(input("Введите опыт работы Самира: "))
pravin_experience: int = int(input("Введите опыт работы Правина: "))
mohit_experience: int = int(input("Введите опыт работы Мохита: "))

# Определение наиболее и наименее опытного сотрудника
most_experienced: int = max(samir_experience, pravin_experience, mohit_experience)
least_experienced: int = min(samir_experience, pravin_experience, mohit_experience)

# Вывод результатов
if samir_experience == most_experienced:
    print("Самир является наиболее опытным сотрудником.")
if pravin_experience == most_experienced:
    print("Правин является наиболее опытным сотрудником.")
if mohit_experience == most_experienced:
    print("Мохит является наиболее опытным сотрудником.")

if samir_experience == least_experienced:
    print("Самир является наименее опытным сотрудником.")
if pravin_experience == least_experienced:
    print("Правин является наименее опытным сотрудником.")
if mohit_experience == least_experienced:
    print("Мохит является наименее опытным сотрудником.")
```

```python
# 5. Даны три угла треугольника. Напишите программу, которая проверит, является
# ли он прямоугольным (один из углов должен составлять 90°).

# Ввод углов треугольника
angle_a, angle_b, angle_c: list[int] = map(int, input().split())

# Проверка на прямоугольный треугольник
if angle_a == 90 or angle_b == 90 or angle_c == 90:
    print("Right triangle")
else:
    if angle_a + angle_b + angle_c == 180 and (
        angle_a + angle_b == 90 or angle_a + angle_c == 90 or angle_b + angle_c == 90
    ):
        print("Right triangle")
    else:
        print("Not right triangle")
```

```python
# 6. Сколько разных трехзначных чисел можно получить, 
# используя цифры 1, 2 и 3 в разных позициях?  
#  Напишите программу для генерации всех таких чисел

count_: int = 0

for i_ in range(1, 4):  # Первая цифра (единицы)
    for j_ in range(1, 4):  # Вторая цифра (десятки)
        for k_ in range(1, 4):  # Третья цифра (сотни)
            if i_ != j_ and j_ != k_ and i_!= k_:  
                number = i_ * 100 + j_ * 10 + k_
                # print(number)# выводит числа
                count_ += 1
print(count_)
```

```python
# 8. Напишите программу для вывода всех простых чисел от 1 до 500.
# https://habr.com/ru/articles/122538/
# for i in range(,501):

# создаем пустой список для хранения простых чисел
lst: list[int] = []
# в k_ будем хранить количество делителей
k_: int = 0
# пробегаем все числа от 2 до N
for i_ in range(1, 500 + 1):
    # пробегаем все числа от 2 до текущего
    for j_ in range(2, i_):
        # ищем количество делителей
        if i_ % j_ == 0:
            k_ = k_ + 1
    # если делителей нет, добавляем число в список
    if k_ == 0:
        lst.append(i_)
    else:
        k_ = 0
# выводим на экран список
print(lst)
```

```python
# 10. Машина в течение срока службы приносит годовой доход в размере 200 000 руб.
# Сама она стоит 1 000 000 руб. в момент покупки и продается за 250 000 руб.
# при утилизации. Вложив ту же сумму в другие инвестиционные инструменты,
# можно заработать 8 % годовых. Каков минимальный срок службы машины, при
# котором она станет более привлекательной по сравнению с альтернативными
# инвестициями?
```

```python
def calculate_profit(years: int) -> float:
    """Вычисляет прибыль от машины за указанный период.

    Аргументы:
        years (int): Количество лет, на которое производится расчет.

    Возвращает:
        float: Разница между прибылью от машины и доходом от инвестиций.
               Если результат положительный, это означает, что машина 
               приносит больше прибыли, чем инвестиции.

    Примечание:
        Прибыль от машины рассчитывается как общий доход за указанный 
        период минус начальная стоимость плюс остаточная стоимость. 
        Доход от инвестиций учитывает сложные проценты.
    """
    
    total_revenue_machine: int = 200000 * years  # общий доход от машины за год
    initial_cost: int = 1000000
    salvage_value: int = 250000
    revenue_investment = (
        initial_cost * (1 + 0.08) ** years
    )  # доход от инвестиций с учетом ставки и сроков с сложным процентом

    profit_machine: float = total_revenue_machine - initial_cost + salvage_value
    profit_investment: float = revenue_investment - initial_cost

    return profit_machine - profit_investment


years = 1
while calculate_profit(years) <= 0:
    years += 1
print(f"Минимальный срок окупаемости машины : {years} лет")
```
