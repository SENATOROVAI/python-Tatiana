# 1.1.3. Настоящее и будущее Data Science

+ [markdown] id="i9AxylmnswrK"<br>
Data Science не закончилось и идет по нескольким направлениям:<br>
1. Искусственный интеллект (ИИ) — это сфера, в которой основное внимание уде­<br>
ляется созданию интеллектуальных машин, способных работать и принимать<br>
решения как человек. Исследования в области ИИ начались около 1936 г., когда<br>
Алан Тьюринг создал первые машины на базе ИИ<br>
2. Машинное обучение— это инструмент для извлечения знаний из данных.<br>
В машинном обучении модели могут обучаться на данных самостоятельно или<br>
поэтапно: обучение с учителем, т. е. на данных, подготовленных человеком, или<br>
обучение без учителя, в котором работа ведется над хаотичными и неорганизо­<br>
ванными данными<br>
3.  Глубокое обучение — это создание многослойных нейронных сетей в областях,<br>
где требуется более продвинутый или быстрый анализ, а традиционное машин­<br>
ное обучение не справляется. Под глубиной понимается наличие более одного<br>
скрытого слоя нейронов в сети, которые проводят математические вычисления.<br>
4. Большие данные — это работа с огромными объемами часто неструктуриро­<br>
ванных данных. Специфика этой сферы — инструменты и системы, способные<br>
выдерживать высокие нагрузки.<br>


+ [markdown] id="6IujSv4gtY8O"<br>

# 1.3.8.1. Сравнение компилятора и интерпретатора<br>
<br>


+ [markdown] id="2bFuVaByUaDG"<br>
1. И компиляторы, и интерпретаторы преобразуют исходный код (текстовые фай­<br>
лы) в машинный язык. Основное их отличие состоит в том, что компилятор<br>
генерирует автономную программу, написанную на машинном коде, а система<br>
интерпретатора выполняет действия, описанные программой высокого уровня.<br>
2. После компиляции программы ее исходный код для запуска больше не требует­<br>
ся. У интерпретируемых программ исходный код требуется для запуска про­<br>
граммы каждый раз.<br>
3. Обычно интерпретируемые программы работают медленнее, чем компили­<br>
руемые.
