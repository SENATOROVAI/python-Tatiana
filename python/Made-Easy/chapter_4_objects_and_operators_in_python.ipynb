
**4.3.4. Изменяемые и неизменяемые объекты**<br>
Значение у некоторых объектов может меняться. Объекты, значение которых мо­<br>
жет меняться, называются изменяемыми, а объекты, значение которых после созда­<br>
ния остается постоянным, называются неизменяемыми. Изменяемость объекта<br>
определяется его типом; например, числа, строки и кортежи неизменны, а словари<br>
и списки — изменяемы.<br>
Некоторые объекты содержат ссылки на другие объекты, и тогда они называются<br>
контейнерами. Примерами контейнеров являются кортежи, списки и словари. Зна­<br>
чения контейнера — это ссылки на объекты.<br>
когда мы го­<br>
ворим о значении контейнера, мы подразумеваем именно значения содержащихся<br>
в нем объектов, а не их адреса. Но говоря об изменяемости контейнера, мы имеем<br>
в виду только адреса содержащихся в нем объектов. если неизменяемый кон­<br>
тейнер (например, кортеж) содержит ссылку на изменяемый объект, значение кон­<br>
тейнера изменится при изменении значения объекта. Однако контейнер по-<br>
прежнему считается неизменяемым, поскольку набор содержащихся в нем объек­<br>
тов не может быть изменен. Получается, что неизменяемость контейнеров — это не<br>
то же самое, что невозможность изменить значение, тут все чуть сложнее.

6. Что делают функции chr () и ord о ? Как они связаны?<br>
* Функции chr() и ord() в Python связаны между собой и позволяют работать с символами в виде их кодовых значений:<br>
<br>
* chr(x): Эта функция принимает целочисленный аргумент x и возвращает строку, содержащую символ, соответствующий этому целому числу в соответствии с текущей кодировкой (обычно UTF-8).<br>


+<br>
Например:

```python
print(chr(65))  # Output: 'A'
print(chr(97))  # Output: 'a'
# -
```

* Эти функции полезны, когда вам нужно преобразовывать между символами и их кодовыми значениями, например, для работы с ASCII или Юникод-кодами.<br>
* Таким образом, chr() и ord() являются взаимно обратными функциями:<br>
<br>

* ord(chr(x)) всегда возвращает исходное целое число x.
